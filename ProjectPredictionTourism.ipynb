```python
#แทนที่ค่า Missing Value ด้วยค่า Means
df = pd.read_csv('tourism_dataset.csv')
means = df.mean()
df_filled = df.fillna(means)
df_filled.to_csv('tourism_datacleaned.csv', index=False)
df = pd.read_csv('tourism_datacleaned.csv')
pd.DataFrame(df)
```

```python
#ตรวจสอบค่า Outlier ใน Dataset
df = pd.read_csv('tourism_datacleaned.csv')

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df < lower_bound) | (df > upper_bound))

print(outliers.sum())
```

```python
#กำจัดค่า Outlier ด้วยค่าเฉลี่ยของแต่ละคอลลัมน์และแสดงผลค่าที่กำจัดแล้ว
df = pd.read_csv('tourism_datacleaned.csv')

columns_to_remove_outliers = [
    'TempApr', 'TempMay', 'TempDec', 'TempFeb',
    'TempNov', 'TempOct', 'TourismApr', 'TourismAug',
    'TourismDec', 'TourismFeb', 'TourismJan', 'TourismJuly',
    'TourismJune', 'TourismMar', 'TourismMay', 'TourismNov',
    'TourismOct', 'TourismSep', 'TourismTotal'
]

Q1 = df[columns_to_remove_outliers].quantile(0.25)
Q3 = df[columns_to_remove_outliers].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df[columns_to_remove_outliers] < lower_bound) | (df[columns_to_remove_outliers] > upper_bound))

column_means = df[columns_to_remove_outliers].mean()

for column in columns_to_remove_outliers:
    df[column] = df[column].apply(lambda x: column_means[column] if x in outliers[column] else x)

df.to_csv('tourism_datacleaned.csv', index=False)
df = pd.read_csv('tourism_datacleaned.csv')
pd.DataFrame(df)
```

```python
# แบ่งข้อมูลเป็นชุดการฝึกและชุดทดสอบ
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print("Train shape: ", X_train.shape)
print("Test shape: ", X_test.shape)
```

```python
# สร้างโมเดล Linear Regression
lr = LinearRegression()

# ฝึกโมเดลโดยใช้ชุดการฝึก
lr.fit(X_train, Y_train)
```

```python
# สร้างโมเดล KNN
knn = KNeighborsRegressor(n_neighbors=5)  # กำหนดจำนวนเพื่อบรรทัด k (เช่น n_neighbors=5)

# ทำการเทรนโมเดล KNN ด้วยข้อมูลการฝึกอบรม
knn.fit(X_train, Y_train)

# ทำการทำนายด้วยข้อมูลทดสอบ
y_knn_pred = knn.predict(X_test)
```

```python
def calculate_metrics(X, Y, model):
    pred = model.predict(X)
    mse = mean_squared_error(Y, pred)
    rmse = mse ** 0.5
    mae = mean_absolute_error(Y, pred)
    r2 = r2_score(Y, pred)

    return rmse, mae, r2
```

```python
lr_rmse, lr_mae, lr_r2 = calculate_metrics(X_test, Y_test, lr)
print(f"RMSE: {lr_rmse}")
print(f"MAE: {lr_mae}")
print(f"R^2: {lr_r2}")
```

```python
knn_rmse, knn_mae, knn_r2 = calculate_metrics(X_test, Y_test, knn)
print(f"RMSE: {knn_rmse}")
print(f"MAE: {knn_mae}")
print(f"R^2: {knn_r2}")
```
